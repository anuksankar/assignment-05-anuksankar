docker run redis
(control c to exit)
docker ps -a
docker rm -f xxxxx
```

The -d option frees up the commandline.
```
docker run -d redis
docker ps -a
docker rm -f xxxxx
```

The --name option allowed me to name the redis process as "redis".

```
docker run -d --name redis redis
docker ps -a
docker rm -f redis
```

The -p option opens the port 6379 for redis

```
docker run -d --name redis -p 6379:6379 redis
docker ps -a
docker rm -f redis
```


#### Creating a docker cluster with redis standalone

As we had already created a directory ~/w205/redis-standalone, I did not / cannot make another directory with the same name.  
So I changed directory to it instead.  Then I copied the example to docker-compose.yml file.

```

From the w205 directory on the droplet, I opened a new Python3 notebook and executed the following commands.  The notebook is saved as anu_sankar_hw5_trip.ipynb

```python
import redis
import pandas as pd

trips=pd.read_csv('trips.csv')

date_sorted_trips = trips.sort_values(by='end_date')

date_sorted_trips.head()

for trip in date_sorted_trips.itertuples():
      print(trip.end_date, '', trip.bike_number, '', trip.end_station_name)

current_bike_locations = redis.Redis(host='redis', port='6379')

current_bike_locations.keys()

for trip in date_sorted_trips.itertuples():
      current_bike_locations.set(trip.bike_number, trip.end_station_name)
      
current_bike_locations.keys()
```
